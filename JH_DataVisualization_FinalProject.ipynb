```python
#****Load the Data****
import pandas as pd
import altair as alt

# Use original full dataset (real header is on row 2)
data = pd.read_csv("NFL_TotalStats1.csv", header=1)
data.head()
```

```python
#****Line Graph to Show Total Yards, Passing Yards, and Rushing Yards Over Time****
# Total Yards
line1 = alt.Chart(data).mark_line().encode(
    x='Year:O',
    y='TYds:Q'
).properties(title='Total Yards')

# Passing Yards
line2 = alt.Chart(data).mark_line(color='orange').encode(
    x='Year:O',
    y='P-Yds:Q'
).properties(title='Passing Yards')

# Rushing Yards
line3 = alt.Chart(data).mark_line(color='green').encode(
    x='Year:O',
    y='R-Yds:Q'
).properties(title='Rushing Yards')


line1 & line2 & line3
```

```python
#****Compare Penalities to Total Yards and Passing Yards****
scat1 = alt.Chart(data).mark_circle().encode(
    x='Pen-Yds:Q',
    y='TYds:Q',
    color=alt.value('blue'),
).properties(title='Penalty Yards vs Total Yards')

scat2 = alt.Chart(data).mark_circle().encode(
    x='Pen-Yds:Q',
    y='P-Yds:Q',
    color=alt.value('orange'),
).properties(title='Penalties vs Passing Yards')

scat1 & scat2
```
